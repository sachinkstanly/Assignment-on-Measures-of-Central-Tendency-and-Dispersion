QUESTION 3: . Who scores the most on average for math, reading and writing based on
● Gender ● Test preparation course

```python
 # Grouping by gender and test preparation course, and calculate the average scores
avg_scores = data_df.groupby(['gender', 'test preparation course']).mean()

# Printing the result
print("Average Scores:")
print(avg_scores)

```

```python
data_df.groupby('gender').mean() # Grouping by gender and calculate the average scores
```

```python
data_df.groupby('test preparation course').mean() # Grouping by test preparation course, and calculate the average scores

```

QUESTION 4: What do you think about the scoring variation for math, reading and writing

```python
# Calculate the standard deviation for math, reading, and writing scores based on gender and test preparation course
variation = data_df.groupby(['gender', 'test preparation course']).std()

# Print the result
print("Scoring Variation:")
print(variation)
```

```python
data_df.groupby('gender').std() # Calculate the standard deviation for math, reading, and writing scores based on gender
```

```python
data_df.groupby('test preparation course').std() # Calculate the standard deviation for math, reading, and writing scores based on test preparation course
```
